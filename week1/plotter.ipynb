```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

# Plotting signals

```python
def complex_exp(samples_array, omega):
    return np.exp(1j * samples_array.astype(complex) * omega)
```

## Define some signal

```python
ns = np.linspace(-2, 2, 500)*np.pi


xn = complex_exp(ns, omega=2) + complex_exp(ns, omega=3)
```

## Plot

```python
fig, ax = plt.subplots(figsize=(18, 6))

# plot in units of pi
ax.plot(ns / np.pi, np.abs(xn))


# set plot 
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
plt.grid()
plt.show()
```
