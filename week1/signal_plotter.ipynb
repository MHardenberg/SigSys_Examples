```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

# Periodicity of signals

```python
def complex_exp(samples_array, omega):
    return np.exp(1j * samples_array.astype(complex) * omega)
```

## Define some signal

```python
ns = np.linspace(0, 2*np.pi, 250)


xn = complex_exp(ns, omega=4) + complex_exp(ns, omega=6)
```

## Plot

```python
fig, ax = plt.subplots(figsize=(18, 6))

# plot in units of pi
ax.plot(ns / np.pi, np.real(xn))


# set plot 
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
plt.grid()
plt.show()
```
