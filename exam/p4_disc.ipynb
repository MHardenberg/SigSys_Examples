```python
# imports
import numpy as np
import matplotlib.pyplot as plt


# define series
def h_fn(n):
    return 2*(np.heaviside(n, 1) - np.heaviside(n-5, 1))

def x_fn(n):
    return np.heaviside(n, 1) - np.heaviside(n-3, 1)

# define ns
n, m = 3, 8
ns = np.arange(-n,m)
xs, hs = x_fn(ns), h_fn(ns)

#plot 
fig, ax = plt.subplots()
ax.stem(ns, hs, linefmt='tab:blue', label='h[n]')
ax.stem(ns, xs, linefmt='tab:orange', label='x{n]')

ax.legend()
plt.savefig("figures/p4_stems.png")
plt.show()
```

```python
ys = np.convolve(xs, hs)
ks = np.arange(-(n + n), m + m - 1)

#plot 
fig, ax = plt.subplots()
ax.stem(ks, ys, linefmt='tab:green', label='y[n]')
ax.set_xlim([-3, 10])
ax.legend()
plt.savefig("figures/p4_conv.png")
plt.show()
```
